```python
from bs4 import BeautifulSoup
from bs4 import SoupStrainer
from bs4.dammit import EncodingDetector
from urllib.parse import urljoin
from urllib.parse import urlparse

parser = "lxml"  # faster than html.parser
http_encoding = r.apparent_encoding  # not r.encoding
html_encoding = EncodingDetector.find_declared_encoding(
    r.content, is_html=True)
encoding = html_encoding or http_encoding

base_url = "https://www.ist.utl.pt/"
for link in BeautifulSoup(content, parser, parse_only=SoupStrainer("a"), from_encoding=encoding):
    if hasattr(link, "href"):
        url = urljoin(base_url, link["href"])
        u = urlparse(url)
        if u.scheme.startswith("http"):
            if rp.can_fetch("*", url):
                print(urldefrag(url).url)  # now we need to add de-duplicated URLs to a processing list...
```

```python
import json
from sseclient import SSEClient as EventSource

url = 'https://stream.wikimedia.org/v2/stream/recentchange'
wiki = 'enwiki'
for event in EventSource(url):
    if event.event == 'message':
        try:
            change = json.loads(event.data)
        except ValueError:
            continue
        if change['wiki'] == wiki:
            print('{0} edited {1}'.format(change["user"], change["meta"]["uri"]))
```
